```python
## 導入 sklearn 數據集裡的digits數據
from sklearn.datasets import load_digits
digits_data = load_digits()
## 列出屬性種類
print(digits_data.keys())
## 裝載樣本數量與特徵數量
n_samples, n_features = digits_data.data.shape
print("樣本數量: ", n_samples)
print('特徵數量: ', n_features)
## 印出data矩陣的形狀大小
print(digits_data.data.shape)
## 印出images矩陣的形狀大小
print(digits_data.images.shape)
import numpy as np
print(np.all(digits_data.images.reshape((1797,64)) == digits_data.data))
## 圖片設定
## subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)，用來調整子圖，有六個參數，值都在0~1之間，left、bottom、right、top圍起來的區域就視子圖的區域，wspace與hspace用來控制子圖間左右、上下的距離
fig = plt.figure(figsize=(6,6))
fig.subplots_adjust(left=0,right=1,bottom=0,top=1,hspace=0.06,wspace=0.06)
## 繪圖, 每張子圖8x8像素
## interpolation 參數可以參考https://matplotlib.org/gallery/images_contours_and_fields/interpolation_methods.html?highlight=interpolation
## cmap 參數可以參考https://blog.csdn.net/lanchunhui/article/details/66972783
# imshow(): 將圖像數據顯示圍圖像, 詳細參數可以參考https://vimsky.com/zh-tw/examples/usage/matplotlib-pyplot-imshow-in-python.html
for i in range(64):
 ## 子圖設定
 ax = fig.add_subplot(8,8,i+1,xticks=[],yticks=[])
 ## 繪入子圖
 ax.imshow(digits_data.images[i], cmap='binary', interpolation='nearest')
 ## 用目標的值，也就是真實值，來標記圖像
 ## text(x,y, string, fontsize = 15, verticalalignment="top",horizontalalignment="right")
 ## x,y: 坐標軸位置、 string: 要說明的文字、 fontsize: 字體大小、 verticalalignment: 垂值對其的方式、 horizontalalignment: 水平齊的方式
 ax.text(0,1,str(digits_data.target[i]))
 
plt.show()
```

```python
## 導入 sklearn 數據庫裡的Iris 鳶尾花數據集
from sklearn.datasets import load_iris
## 載入數據
iris_data = load_iris()
## 列出有哪些屬性
print(iris_data.keys())
## 樣本總數與特徵(維度)數
n_samples, n_features = iris_data.data.shape
print('樣本數量: ', n_samples)
print('特徵數量: ', n_features)
## 導入數據處理套件
import numpy as np
## 列出目標種類名稱
print(iris_data.target_names)
## 種類的各別數量
print(np.bincount(iris_data.target))
## 導入可視化套件
import matplotlib.pyplot as plt
## 繪圖
## 選擇特徵欄位的前面兩個特徵來繪圖, 由於視2維圖，所以只取兩個特徵藍位資料
x_index = 0
y_index = 1
## 選三個顏色分別代表三種鳶尾花種類
colors = ['blue','red','yellow']
## label 為0,1,2, 對應的顏色是blue,red,yellow
for label,color in zip(range(len(iris_data.target_names)),colors):
 ## print(label)
 ## print(color)
 plt.scatter(iris_data.data[iris_data.target == label, x_index], iris_data.data[iris_data.target == label, y_index],
 label = iris_data.target_names[label],
 c = color)
## 背景顏色
ax = plt.gca()
ax.set_facecolor('pink')
plt.xlabel(iris_data.feature_names[x_index])
plt.ylabel(iris_data.feature_names[y_index])
plt.legend(loc='best')
plt.show()
## 顯示一個特徵值欄位在三種種類下的數值
## 我們選特徵值欄位的第三個特徵值
x_index = 2
## 選三個顏色分別代表三種鳶尾花種類
colors = ['blue','red','yellow']
for label,color in zip(range(len(iris_data.target_names)),colors):
 plt.hist(iris_data.data[iris_data.target == label, x_index],
 label = iris_data.target_names[label],
 color = color)
plt.xlabel(iris_data.feature_names[x_index])
plt.legend(loc='best')
plt.show()
```

```python
## 導入樣本生成器套件
from sklearn.datasets.samples_generator import make_classification
## 產生用以分類的數據集
X, y = make_classification(n_samples=400, n_features=10, n_informative = 6, n_redundant = 2, n_repeated = 2, n_classes = 3, n_clusters_per_class = 2, scale = 1.0, random_state =2)
## 印出數據
for data, target in zip(X,y):
 print(target, end=": ")
 print(data)
## 根據第一個與第三個特徵值欄位來繪圖
plt.scatter(X[:, 0], X[:, 2], marker ='o', c=y, s=26, edgecolor='k')
plt.title("Make Classification")
plt.show()
```
